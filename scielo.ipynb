```python
import re
from langdetect import detect


qtd_pag = driver.find_element(By.XPATH, '/html/body/section/div/div/div[1]/div[2]/div[1]/div[2]').text

print(qtd_pag)

numero_total = re.findall(r'\d+', qtd_pag)


lista_links = []
p = 1
total_p = 1#int(numero_total[0])
while p <= total_p:
    try:
        for i in range(1, 51):                            
                titulo = driver.find_element(By.XPATH, f'/html/body/section/div/div/div[1]/div[2]/div[3]/div[{i}]/div[2]/div[1]/a/strong')
                idioma = detect(titulo.text)             
                if idioma == 'pt':
                    link = driver.find_element(By.XPATH, f'/html/body/section/div/div/div[1]/div[2]/div[3]/div[{i}]/div[2]/div[1]/a').get_attribute('href')                 
                    lista_links.append(link)         
        
        if p == 1:
            driver.find_element(By.XPATH, '/html/body/section/div/div/div[1]/div[2]/div[1]/div[2]/a').click()
        else:
            driver.find_element(By.XPATH, '/html/body/section/div/div/div[1]/div[2]/div[1]/div[2]/a[2]').click()
            
        p += 1
    except:
        print('Fim da busca sobre as páginas')
        break
                                    
lista_links
```

```python
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
import datetime

def obter_titulo(driver):
    try:
        return driver.find_element(By.CLASS_NAME, 'article-title').text
    except NoSuchElementException:
        return None

def obter_resumo(driver):
    try:
        div_resumo = driver.find_element(By.XPATH, '//*[@id="articleText"]/div[1]')
        return div_resumo.find_element(By.TAG_NAME, 'p').text
    except NoSuchElementException:
        return None

def obter_texto_completo(driver):
    try:
        div_texto = driver.find_element(By.XPATH, '//div[@data-anchor="Text"]')
    except NoSuchElementException:
        try:
            div_texto = driver.find_element(By.XPATH, '//*[@id="articleText"]/div[3]')
        except NoSuchElementException:
            return None
    
    paragrafos = div_texto.find_elements(By.TAG_NAME, 'p')
    textos_paragrafos = [texto.text for texto in paragrafos]
    return '\n'.join(textos_paragrafos)

def obter_referencias(driver):
    try:
        div_li = driver.find_element(By.CLASS_NAME, 'ref-list')
        li = div_li.find_elements(By.TAG_NAME, 'li')
        referencias = [elemento.text for elemento in li]
    except NoSuchElementException:
        return None
    
    return referencias

def obter_datas_publicacao(driver):
    try:
        ul_datas_publi = driver.find_element(By.CLASS_NAME, 'articleTimeline')
        li_datas_publi = ul_datas_publi.find_elements(By.TAG_NAME, 'li')
        datas_publi = [data.text for data in li_datas_publi]
    except NoSuchElementException:
        return None
    
    return datas_publi

def obter_autores_detalhes(driver):
    try:
        botao_autores = driver.find_element(By.CLASS_NAME, 'outlineFadeLink')            
        driver.execute_script("arguments[0].click();", botao_autores)
        div_autores = driver.find_elements(By.CLASS_NAME, 'tutors')                     
        autores = [div.find_element(By.TAG_NAME, 'strong').get_attribute('textContent') for div in div_autores]
        detalhes = [div.find_element(By.TAG_NAME, 'span').get_attribute('textContent') for div in div_autores] 
    except NoSuchElementException:
        return None, None   
    
    return autores, detalhes

dados_lista = []
contador_links = 1

for link in lista_links:
    driver.get(link)

    titulo = obter_titulo(driver)
    resumo = obter_resumo(driver)
    texto_completo = obter_texto_completo(driver)
    referencias = obter_referencias(driver)
    datas_publi = obter_datas_publicacao(driver)
    autores, detalhes = obter_autores_detalhes(driver)

    dados = {
        'Título': titulo,
        'Resumo': resumo,
        'Texto Completo': texto_completo,
        'Referências': referencias,
        'Datas de Publicação': datas_publi,
        'Autores': autores,
        'Detalhes': detalhes,
        'Link': link,
        'Data de Coleta': datetime.datetime.now(),
        'Portal': 'Scielo.org'
    }

    dados_lista.append(dados)

    print('Links percorridos: ', {contador_links})
    contador_links += 1

df = pd.DataFrame(dados_lista)

driver.quit()

display(df.shape)
```
